 # <u>Exploring Polynomial Bases </u>

In this notebook, we'll explore the concept of polynomials and different polynomial bases. Specifically, we will be looking at Vandermonde, Legendre and Chebyshev bases. Let us first start by defining a polynomial

 ### <u>Polynomial Definition</u>

A polynomial of degree $n$ is defined as:

$$
P(x) = a_nx^n + a_{n-1}x^{n-1} + \ldots + a_1x + a_0
$$

where:
- $P(x)$ is the polynomial function,
- $a_n, a_{n-1}, \ldots, a_1, a_0$ are the coefficients of the polynomial, and
- $n$ is the degree of the polynomial.

For example, let's consider a quadratic polynomial:

$$
P(x) = 3x^2 - 8x + 5
$$

Here,
- $a_2 = 3$,
- $a_1 = -8$,
- $a_0 = 5$, and
- $n = 2$.

Before moving on to polynomial bases let us look at the graph of this polynomial and discuss what we mean by the root of a polynomial.

### <u><b>Polynomial Basis</u></b>

A polynomial basis is a set of polynomials that form a basis for a vector space of polynomials. In other words, any polynomial in the vector space can be expressed as a linear combination of the polynomials in the basis.

For example, the set $\{1, x, x^2, x^3\}$ forms a polynomial basis for polynomials of degree at most 3. Any polynomial $P(x)$ of degree at most 3 can be expressed as:

$$
P(x) = a_0 \cdot 1 + a_1 \cdot x + a_2 \cdot x^2 + a_3 \cdot x^3
$$

where $a_0, a_1, a_2, a_3$ are coefficients. This representation demonstrates that any polynomial of degree at most 3 can be uniquely represented as a linear combination of the polynomials in the basis.

Similarly, for polynomials of degree $n$, the set $\{1, x, x^2, \ldots, x^n\}$ forms a polynomial basis.

### <u><b>Example</u></b>

Consider the polynomial $P(x) = 2x^3 - 3x^2 + 5x - 1$. To express $P(x)$ using the polynomial basis $\{1, x, x^2, x^3\}$, we identify the coefficients:

$$
\begin{align*}
a_0 &= -1 \\
a_1 &= 5 \\
a_2 &= -3 \\
a_3 &= 2
\end{align*}
$$

Therefore, $P(x)$ can be expressed as:

$$
P(x) = -1 \cdot 1 + 5 \cdot x - 3 \cdot x^2 + 2 \cdot x^3
$$

This demonstrates that $P(x)$ can be represented as a linear combination of the polynomials in the polynomial basis.

Let us now look at some of the frequently used polynomial bases starting with the Vandermonde basis.


### <u>Vandermonde Basis in Polynomial Space</u>

The Vandermonde matrix is a matrix of the form:

$$
V = \begin{bmatrix}
1 & x_1 & x_1^2 & \cdots & x_1^n \\
1 & x_2 & x_2^2 & \cdots & x_2^n \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_m & x_m^2 & \cdots & x_m^n \\
\end{bmatrix}
$$

where $x_1, x_2, \ldots, x_m$ are distinct real or complex numbers. It constructs a basis for the space of polynomials of degree at most $n$.

### <u>Vandermonde Basis</u>

The Vandermonde basis for polynomials of degree $n$ is defined as the set of polynomials:

$$
\{1, x, x^2, \ldots, x^n\}
$$

This basis is derived from the columns of the Vandermonde matrix $V$, where $x$ takes on distinct values.

### <u>Example: Degree 3 Polynomial</u>

Consider the degree 3 polynomial $P(x) = 2x^3 + 3x^2 + 4x + 1$. We can express $P(x)$ using the Vandermonde basis by constructing the Vandermonde matrix with four distinct values of $x$, say $x_1, x_2, x_3, x_4$. For simplicity, let's choose $x_1 = 1$, $x_2 = 2$, $x_3 = 3$, and $x_4 = 4$. 

The Vandermonde matrix $V$ would be:

$$
V = \begin{bmatrix}
1 & 1 & 1 & 1 \\
1 & 2 & 4 & 8 \\
1 & 3 & 9 & 27 \\
1 & 4 & 16 & 64 \\
\end{bmatrix}
$$

To find the coefficients of the polynomial $P(x)$ in the Vandermonde basis, we solve the linear system $V \mathbf{a} = \mathbf{b}$, where $\mathbf{a} = [a_0, a_1, a_2, a_3]$ are the coefficients of the polynomial and $\mathbf{b} = [1, 4, 3, 2]$ are the function values of $P(x)$ at $x_1, x_2, x_3, x_4$ respectively.

Solving this system, we get $\mathbf{a} = [1, 4, 3, 2]$, which are the coefficients of $P(x)$ in the Vandermonde basis. Therefore, the polynomial $P(x)$ can be expressed as:

$$
P(x) = 1 \cdot 1 + 4 \cdot x + 3 \cdot x^2 + 2 \cdot x^3
$$

This demonstrates that the polynomial $P(x)$ can be represented as a linear combination of the polynomials in the Vandermonde basis.

Note that we need not solve the whole system to find the coefficients. The values are easily obtainaible just by observation. This method is just to demonstrate how we can use the Vandermonde matrix for polynomial fitting in case we want to fit a polynomial of degree n given n+1 points. In case we have more than n+1 points, we obtain a least square fit using the ideas of LSP(Least squares problem). We end the discussion on the Vandermonde basis by looking at this explanation in a bit more detail


### <u>Fitting a Polynomial using Vandermonde Basis and least square method</u>

In this Python code demonstration, we are fitting a 4-degree polynomial to 10 randomly generated data points. Here's a step-by-step explanation:

1. **Data Generation**: We generate 10 random data points using a polynomial function with some added noise for variability.

2. **Vandermonde Matrix**: We construct the Vandermonde matrix $ V $ using NumPy's `np.vander()` function. This matrix will be used to form the basis for our polynomial fitting. The Vandermonde matrix $( V )$ is represented as:

$$
V = \begin{bmatrix}
1 & x_1 & x_1^2 & x_1^3 & x_1^4 \\
1 & x_2 & x_2^2 & x_2^3 & x_2^4 \\
\vdots & \vdots & \vdots & \vdots & \vdots \\
1 & x_{10} & x_{10}^2 & x_{10}^3 & x_{10}^4 \\
\end{bmatrix}
$$

3. **Least Squares Solution**: We solve the linear least squares problem using $ V \mathbf{a} = \mathbf{y} $ to find the coefficients $ \mathbf{a} $ of the polynomial that best fits the data points. Here, $ \mathbf{a}  $ are the coefficients of the polynomial, and $ \mathbf{y} $ are the data points.

4. **Fitted Curve**: We generate points for the fitted curve using the coefficients obtained from the least squares solution and plot it along with the original data points.

The resulting plot illustrates how the 4-degree polynomial closely fits the given data points, demonstrating the effectiveness of using the Vandermonde basis for polynomial fitting.



```python
# Generate some random data points
np.random.seed(0)
x = np.linspace(0, 10, 10)
y = 2 * x**3 - 3 * x**2 + 4 * x + 1 + np.random.normal(0, 5, size=10)

# Define the degree of polynomial to fit
degree = 4

# Construct the Vandermonde matrix
V = np.vander(x, degree + 1, increasing=True)

# Solve the linear least squares problem to find the coefficients
coefficients, residuals, _, _ = np.linalg.lstsq(V, y, rcond=None)

# Generate points for the fitted curve
x_fit = np.linspace(0, 10, 100)
y_fit = np.polyval(coefficients[::-1], x_fit)

# Plot the original data points and the fitted curve
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', label='Data Points')
plt.plot(x_fit, y_fit, 'r-', label='Least Square Fit Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Square Fit of 4-degree Polynomial')
plt.legend()
plt.grid(True)
plt.show()
```

### <u>Orthogonal Bases for Polynomials</u>

In polynomial spaces, orthogonal bases are sets of polynomials where each pair of polynomials in the set is orthogonal to each other under a given inner product. Orthogonal bases have several properties that make them useful for various applications, including numerical computations and signal processing.

#### <u>Orthogonal Basis</u>

An orthogonal basis for a polynomial space is a set of polynomials where each pair of polynomials is orthogonal to each other under a specified inner product. Mathematically, if  $\langle p_i, p_j \rangle $ represents the inner product of two polynomials $ p_i $ and $ p_j $, then an orthogonal basis satisfies:

$$
\langle p_i, p_j \rangle = 0 \quad \text{for } i \neq j
$$

where $ p_i $ and $ p_j $ are polynomials in the orthogonal basis.

#### Example: Polynomials with Two Degrees

Consider the space of polynomials with degrees up to 2 (quadratic polynomials). An orthogonal basis for this space can be constructed using the Legendre polynomials, denoted as $ P_0(x) $, $ P_1(x) $, and $ P_2(x) $These polynomials satisfy:

$$
\int_{-1}^{1} P_i(x) P_j(x) \, dx = \begin{cases} 
0 & \text{if } i \neq j \\
c_i & \text{if } i = j 
\end{cases}
$$

where $ c_i $ is a constant.

For instance, the first three Legendre polynomials are:

$$
\begin{align*}
P_0(x) &= 1 \\
P_1(x) &= x \\
P_2(x) &= \frac{1}{2}(3x^2 - 1)
\end{align*}
$$

These polynomials form an orthogonal basis for the space of quadratic polynomials. One can verify by doing the relevant integrals and confirming they are equal to 0. We will look more into Legendre polynomials in the next section.


### <u>Legendre Polynomials and Orthogonal Bases</u>

Legendre polynomials are a sequence of orthogonal polynomials that arise in solving various mathematical problems, particularly in the context of orthogonal bases for polynomial spaces.

#### <u>Legendre Polynomials</u>

The Legendre polynomials $ P_n(x) $ are defined on the interval $ [-1, 1] $ and are orthogonal with respect to the $ L^2 $ norm, which is the integral from $ -1 $ to $ 1 $ of the square of the function. The general formula for Legendre polynomials is given by the Rodrigues' formula:

$$
P_n(x) = \frac{1}{2^n n!} \frac{d^n}{dx^n} \left[ (x^2 - 1)^n \right]
$$

The first few Legendre polynomials are:

$$
\begin{align*}
P_0(x) &= 1 \\
P_1(x) &= x \\
P_2(x) &= \frac{1}{2}(3x^2 - 1) \\
P_3(x) &= \frac{1}{2}(5x^3 - 3x) \\
&\vdots
\end{align*}
$$

The general formula for the nth degree Legendre polynomial is:

$$
P_n(x) = \frac{1}{2^n n!} \frac{d^n}{dx^n} \left[ (x^2 - 1)^n \right]
$$

#### Normed Space

The space of polynomials equipped with the $ L^2 $ norm, denoted as $ L^2([-1, 1]) $, is a normed vector space. The $ L^2 $ norm of a polynomial $ f(x) $ on the interval $ [-1, 1] $ is defined as:

$$
\| f \|_2 = \left( \int_{-1}^{1} |f(x)|^2 \, dx \right)^{1/2}
$$

This norm measures the "length" or "size" of a polynomial in the space, and it satisfies the properties of a norm, such as positivity, homogeneity, and the triangle inequality.

Legendre polynomials form an orthogonal basis for $ L^2([-1, 1]) $, meaning that they are a complete set of orthogonal functions that span the space of polynomials up to a certain degree.

We will later look at a general method to find the nth degree Legendre polynomial without using the above formulae. For the time being let us look at the graphs of various Legendre polynomials for visualization. Note that the recurrence relation used for generating the Legendre polynomial can be easily found by differentiating the recurrence relation twice.


```python
# Define Legendre polynomials
def legendre(n, x):
    if n == 0:
        return np.ones_like(x)
    elif n == 1:
        return x
    else:
        return ((2*n - 1) * x * legendre(n-1, x) - (n - 1) * legendre(n-2, x)) / n

# Generate x values
x = np.linspace(-1, 1, 400)

# Plot Legendre polynomials up to degree 5
plt.figure(figsize=(10, 6))
for n in range(5):
    plt.plot(x, legendre(n, x), label=f'L_{n}(x)')

# Add labels and legend
plt.title('Legendre Polynomials')
plt.xlabel('x')
plt.ylabel('L_n(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

### Chebyshev Polynomials and Orthogonal Bases

Chebyshev polynomials are a sequence of orthogonal polynomials that play a significant role in various mathematical and scientific fields. They are particularly useful for approximation theory, numerical analysis, and solving differential equations.

#### Chebyshev Polynomials

The Chebyshev polynomials $ T_n(x) $ are defined on the interval $ [-1, 1] $ and form an orthogonal basis for polynomial spaces under a specific norm. The general formula for Chebyshev polynomials is given by:

$$
T_n(x) = \cos(n \cos^{-1}(x))
$$

The first few Chebyshev polynomials are:

$$
\begin{align*}
T_0(x) &= 1 \\
T_1(x) &= x \\
T_2(x) &= 2x^2 - 1 \\
T_3(x) &= 4x^3 - 3x \\
&\vdots
\end{align*}
$$

#### Orthogonal Basis and Inner Product

Chebyshev polynomials form an orthogonal basis for the space of polynomials with degrees up to $ n $ under a specific norm. The inner product of any two Chebyshev polynomials $ T_i(x) $ and $ T_j(x) $ is zero if $ i \neq j $ and nonzero if $ i = j $. Mathematically, this can be expressed as:

$$
\langle T_i, T_j \rangle = \int_{-1}^{1} T_i(x) T_j(x) \frac{1}{\sqrt{1-x^2}} \, dx
 = \begin{cases} 
0 & \text{if } i \neq j \\
c_i & \text{if } i = j 
\end{cases}
$$

where $ c_i $ is a constant that depends on the degree of the Chebyshev polynomial $ T_i(x) $.

#### Normed Space

The space of polynomials equipped with the norm induced by the inner product of Chebyshev polynomials, denoted as $ L^2([-1, 1]) $, is a normed vector space. The norm of a polynomial $ f(x) $ on the interval $ [-1, 1] $ is defined as:

$$
\| f \|_2 = \left( \int_{-1}^{1} |f(x)|^2 \frac{1}{\sqrt{1-x^2}} \, dx \right)^{1/2}
$$

This norm measures the "length" or "size" of a polynomial in the space, and it satisfies the properties of a norm, such as positivity, homogeneity, and the triangle inequality.

Chebyshev polynomials play a crucial role in various areas of mathematics, engineering, and physics, due to their orthogonality and approximation properties. As for Legendre polynomials let us take a quick look at the plots of the first few Chebyshev polynomials.


```python
# Generate x values
x = np.linspace(-1, 1, 400)

# Plot Chebyshev polynomials up to degree 5
plt.figure(figsize=(10, 6))

# T_0(x) = 1
plt.plot(x, np.ones_like(x), label='$T_0(x)$')

# T_1(x) = x
plt.plot(x, x, label='$T_1(x)$')

# T_2(x) = 2x^2 - 1
plt.plot(x, 2*x**2 - 1, label='$T_2(x)$')

# T_3(x) = 4x^3 - 3x
plt.plot(x, 4*x**3 - 3*x, label='$T_3(x)$')

# T_4(x) = 8x^4 - 8x^2 + 1
plt.plot(x, 8*x**4 - 8*x**2 + 1, label='$T_4(x)$')

# Add labels and legend
plt.title('Chebyshev Polynomials')
plt.xlabel('x')
plt.ylabel('$T_n(x)$')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

# Gram-Schmidt Process

The Gram-Schmidt process is a method used to orthogonalize a set of vectors in an inner product space. It is particularly useful for constructing an orthogonal basis from a linearly independent set of vectors.

## Process Explanation

Given a set of linearly independent vectors $ \{v_1, v_2, \ldots, v_n\} $ in an inner product space, the Gram-Schmidt process constructs an orthogonal set $ \{u_1, u_2, \ldots, u_n\} $ such that $ \text{span}\{u_1, u_2, \ldots, u_k\} = \text{span}\{v_1, v_2, \ldots, v_k\} $ for all $ k = 1, 2, \ldots, n $.

The process involves iteratively constructing orthogonal vectors as follows:

1. **Initialization**: Let $ u_1 = v_1 $.
2. **Orthogonalization**: For $ k = 2, 3, \ldots, n $,
   - Compute $ u_k = v_k - \sum_{i=1}^{k-1} \frac{\langle u_i, v_k \rangle}{\|u_i\|^2} u_i $.
3. **Normalization**: Normalize $ u_k $ to obtain an orthogonal set $ \{u_1, u_2, \ldots, u_n\} $.

Using this method and the appropriate norm for the correspinding Legendre and Chebysev polynomials we can construct Legendre and Chebysev bases starting with the constant polynomial 1 and taking the set v as the Vandermonde basis.
